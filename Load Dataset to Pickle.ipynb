```python
DATASET_DIR = '/content/drive/My Drive/Kuliah/Semester 6/Deep Learning/Proyek Final DL/Baru/Dataset/'
```

```python
def Dataset_loader(DIR, RESIZE=224):
    IMG = []
    for IMAGE_NAME in tqdm(os.listdir(DIR)):
        PATH = os.path.join(DIR,IMAGE_NAME)
        img = cv2.imread(PATH)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = cv2.resize(img, (RESIZE,RESIZE))
        IMG.append(np.array(img))
    return np.array(IMG)
```

```python
def plot_any(data, title, rows, cols):
  fig=plt.figure(figsize=(3*cols, 3*rows))

  for i in range(1, cols*rows+1):
    ax = fig.add_subplot(rows, cols, i)
    ax.title.set_text(title)
    plt.imshow(data[i-1])
  plt.show()  
```

```python
seen_hands = ['Andrew', 'Enrico', 'Henry', 'Greg', 'Nic', 'Kaggle']
unseen_hands = ['Tomy']
classes = ['NOTHING'] + [chr(i) for i in range(65,91)]
```

```python
train_data = []
seen_test_data = []
unseen_test_data = []

for label in classes:
  print('Loading label', label)
  label_train_data = []
  label_seen_test_data = []
  label_unseen_test_data = []
  for seen in seen_hands:
    print('  Loading data', seen)
    data = Dataset_loader(DATASET_DIR+seen+'/'+label)
    label_train_data.extend(data[:10])
    label_seen_test_data.extend(data[10:])
  for unseen in unseen_hands:
    print('  Loading data', unseen)
    data = Dataset_loader(DATASET_DIR+unseen+'/'+label)
    label_unseen_test_data.extend(data)
  train_data.append(label_train_data)
  seen_test_data.append(label_seen_test_data)
  unseen_test_data.append(label_unseen_test_data)
```

```python
print(len(train_data[0]), len(seen_test_data[0]), len([0]))
```

```python
fileName = DATASET_DIR + 'data_train/raw'
fileObject = open(fileName, 'wb')
pkl.dump(train_data, fileObject)
fileObject.close()

fileName = DATASET_DIR + 'data_test_seen/raw'
fileObject = open(fileName, 'wb')
pkl.dump(seen_test_data, fileObject)
fileObject.close()

fileName = DATASET_DIR + 'data_test_unseen/raw'
fileObject = open(fileName, 'wb')
pkl.dump(unseen_test_data, fileObject)
fileObject.close()
```
